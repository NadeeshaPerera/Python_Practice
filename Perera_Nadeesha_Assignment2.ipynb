```python
# create a copy of the data frame
df2 = df.copy()
df2.head(5)
```

#### Treat missing values

```python
# The lgID is NaN for the league NA. Need to correct this
df2['lgID'].fillna('NA', inplace = True)
df2.head(5)
```

```python
# The NaN in the other columns will be replaced with 0
df2.fillna(0, inplace = True)
df2.head(5)
```

#### Calculate OBP (On Base Percentage), PA (Plate Appearance) and BA (Batting Average)

```python
# OBP = (Hits + Walks + Hit by Pitch) / (At Bats + Walks + Hit by Pitch + Sacrifice Flies)
df2['OBP'] = (df2['H'] + df2['BB'] + df2['HBP'])/(df2['AB'] + df2['BB'] + df2['HBP'] + df2['SF'])
df2.head(5)
```

```python
# Let PA = AB + BB + HBP + SH + SF
df2['PA'] = (df2['AB'] + df2['BB'] + df2['HBP'] + df2['SH'] + df2['SF'] )
df2.head(5)
```

```python
# BA is the number of hits gotten by a player divided by his number of at bats
df2['BA'] = df2['H']/df2['AB']
df2.head(5)
```

3)	How many players have hit 40 or more HRs in one single season? (Number only)

```python
# Group by Player and year
df3 = df2.groupby(['playerID', 'yearID']).sum()
# Select the rows of players that have played 40 or more HR's in a single year
df3 = df3[df3['HR'] >= 40]
# Get the count of the playerID column
df3 = df3.reset_index()
df3.playerID.nunique()
```

24)	 Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe, first name, last name, HRs, season)

25)	Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe, first name, last name, season they eclipsed more than 300 HRs)

```python
df25 = df2.copy()
df25['age'] = df25['yearID'] - df25['birthYear']
df25 = df25.sort_values(['playerID', 'yearID'])
df25['cum_HR'] = df25.groupby(['playerID'])['HR'].cumsum()
df25 = df25[df25['cum_HR'] >= 300]
df25 = df25.groupby(['playerID', 'nameFirst', 'nameLast']).first().reset_index()
df25 = df25[['playerID', 'nameFirst', 'nameLast', 'yearID', 'age', 'HR', 'cum_HR']].sort_values('age')
df25[['nameFirst', 'nameLast', 'yearID']].head(5)
```
